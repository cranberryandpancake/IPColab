```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as pit
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
pit.imshow (gimg1, cmap= 'gray')
#cv2_imshow(gimg1)
```

```python
gx1 = np.array([[-1,0],[0,1]], dtype=int)
gy1 = np.array([[0,-1],[1,0]], dtype=int)
```

```python
gx2 = np.array([[-1,0,0],[-2,0,2],[-1,0,1]], dtype=int)
gy2 = np.array([[1,2,1],[0,0,0],[-1,-2,-1]], dtype=int)
```

```python
x=cv2.filter2D(gimg1,-1,gx1)
y=cv2.filter2D(gimg1,-1,gy1)
```

```python
absX =cv2.convertScaleAbs(x)
absY =cv2.convertScaleAbs(y)
```

```python
outImg = cv2.addWeighted(absX,1,absY,1,0)
pit.imshow(outImg, cmap='gray')
```

```python
gBIn = np.zeros(gimg1.shape, dtype=int)
threshold = 40
for i in range(gimg1.shape[0]):
  for j in range(gimg1.shape[1]):
    if gimg1[i,j] > threshold:
      gBIn[i,j] = 255

pit.imshow(gBIn, cmap='gray')
```
